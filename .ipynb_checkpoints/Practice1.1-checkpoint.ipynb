## 1. Data Loading

```python
import numpy as np
from tensorflow.keras.utils import to_categorical
import keras 
from keras import layers
from keras.datasets import fashion_mnist
```

```python
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()
```

## 2. Preprocessing

```python
print(f"Train images: {train_images.shape}")
print(f"Train labels: {train_labels.shape}")
print(f"Test images: {test_images.shape}")
print(f"Test labels: {test_labels.shape}")
```

#### Data flattening

```python
train_images = train_images.reshape((60000, 28*28))
test_images = test_images.reshape((10000, 28*28))
```

#### Converting [0,255] integers to [0,1] floats

```python
train_images = train_images.astype('float32') / 255
test_images = test_images.astype('float32') / 255
```

#### One hot encoding

```python
train_labels = to_categorical(train_labels)
#test_labels = to_categorical(test_labels)
```

#### Validation data separation

```python
val_images = train_images[:10000]
val_labels = train_labels[:10000]
train_images = train_images[10000:]
train_labels = train_labels[10000:]
```

## 3. Neural Network Setup

We import some relevant libraries and functions to study the model's results:

```python
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report

def plot(axis, train, validation, title):
    epochs = range(1, len(train) + 1)
    axis.plot(epochs, train, 'b-o', label='Training ' + title)
    axis.plot(epochs, validation, 'r--o', label='Validation '+ title)
    axis.set_title('Training and validation ' + title)
    axis.set_xlabel('Epochs')
    axis.set_ylabel(title)
    axis.legend()    

def multiplot(history):
    fig, axes = plt.subplots(1, 2)
    fig.set_figwidth(11)
    plot(axes[0], history.history['loss'], history.history['val_loss'], 'loss')
    plot(axes[1], history.history['accuracy'], history.history['val_accuracy'], 'accuracy')
    plt.show()

target_names = ['T-shirt/top', 
                'Trouser', 
                'Pullover', 
                'Dress', 
                'Coat', 
                'Sandal', 
                'Shirt', 
                'Sneaker', 
                'Bag', 
                'Ankle boot']
```

First, we stablish some base model to apply changes to. We will tweak every hyperparameter to see how the results change, in order to contruct a more refined model from the modifications that brought the best results.

```python
def build_model(n_layers = 2, 
                n_per_layer = 50, 
                layer_act_func = 'sigmoid', 
                output_act_func = 'softmax',
                learning_rate = 0.001,
                loss_function = "categorical_crossentropy",
                metrics = ["accuracy"],
                epochs = 25,
                batch_size = 512):
    # Neural network setup
    _model = keras.Sequential(name='model')
    _model.add(layers.Input(shape=(28*28, )))
    for _lay in range(n_layers):
        _model.add(layers.Dense(n_per_layer, activation = layer_act_func)) 
    _model.add(layers.Dense(10, name = 'output', activation = output_act_func))
    _model.summary()
    # Training setup
    _opt = keras.optimizers.Adam(learning_rate = learning_rate)
    _model.compile(optimizer = _opt, loss = loss_function, metrics = metrics)
    _history = _model.fit(train_images,
                          train_labels,
                          epochs = epochs,
                          batch_size = batch_size,
                          validation_data = (val_images, val_labels))
    return _model, _history
```

### **Setup 1: base model**

For the base model, we have chosen a neural network with only two layers, as only one would result in a very simple model. We will have dense layer with 50 nodes each and sigmoid activation functions. The softmax function was chosen for the output layer in orther to get a vector result that sum up to one, as this values can be understood as the probability of the image to correspond to eaxh label. For the training we have chosen a very standard configuration, but with a big batch size.

Structural hyperparameters:
- Layers: 2
- Neurons per layer: 50
- Layer activation function: sigmoid
- Output activation function: softmax

Learning hyperparameters:
- Optimizer: adam
- Loss function: categorical_crossentropy
- Learning rate: 0.001 
- Epoch: 25
- Batch size: 512

```python
model1, history1 = build_model()
```

```python
predicted_values1 = model1.predict(test_images)
predicted_classes1 = np.argmax(predicted_values1, axis = 1)
report1 = classification_report(test_labels, predicted_classes1, target_names = target_names)
multiplot(history1)
print ("Classification Report Setup 1:")
print (report1)
```

### **Setup 3: more layers**

Structural hyperparameters:
- **Layers: 7**
- Neurons per layer: 50
- Layer activation function: sigmoid
- Output activation function: softmax

Learning hyperparameters:
- Optimizer: adam
- Loss function: categorical_crossentropy
- Learning rate: 0.001 
- Epoch: 25
- Batch size: 512

```python
model3, history3 = build_model(n_layers = 7)
```

```python
predicted_values3 = model3.predict(test_images)
predicted_classes3 = np.argmax(predicted_values3, axis =1)
report3 = classification_report(test_labels, predicted_classes3, target_names = target_names)
multiplot(history3)
print ("Classification Report Setup 3:")
print (report3)
```

**Observation**: increasing the networtks depth seems to wornsen de result. This might be caused beacause a lower depth is enough to express the problems relations and increasing the number of layers just makes the training longer, as more parameters need to be trained. We don't see that adding layers to the model makes sense in this case, as the results for just 2 seem to be good enough.

### **Setup 5: more learnig rate**

Learning hyperparameters:

### **Setup 7: less batch size, more epochs**

Structural hyperparameters:
- Layers: 2

Learning hyperparameters:

```python
predicted_values7 = model7.predict(test_images)
predicted_classes7 = np.argmax(predicted_values7, axis = 1)
report7 = classification_report(test_labels, predicted_classes7, target_names = target_names)
multiplot(history7)
print ("Classification Report Setup 7:")
print (report7)
```

**Observation**: reducing the batch size seems to slightly increase the accuracy of the model. An the other hand, the batch size is the number of samples that we use for each weight update in each epoch, so reducing the batch size makes the training longer, as it makes more updates. 

## 4. Regularization

#### Early stopping

```python
from keras.callbacks import EarlyStopping
early_stopping = EarlyStopping(monitor='val_loss', mode="min", patience=5) # Usar val_accuracy (max)
```

#### Dropout, batch normalization, weight regularization and initialization

```python
from keras import regularizers, initializers
dropout_per = 0.1 # Dropout percentage per dense layer
reg = regularizers.l2(0.01) # L2 regularizer
init = initializers.GlorotUniform() # GlorotUniform initializer
```

```python
model_reg = keras.Sequential(name='fashion_mnist')
model_reg.add(layers.Input(shape=(28*28, )))
model_reg.add(layers.Dense(50, kernel_regularizer = reg, name='hidden_1', activation='sigmoid', kernel_initializer = init))
model_reg.add(layers.Dropout(dropout_per)) # Dropout layer
model_reg.add(layers.BatchNormalization()) # Batch normalization layer
model_reg.add(layers.Dense(50, kernel_regularizer = reg, name='hidden_2',activation='sigmoid', kernel_initializer = init))
model_reg.add(layers.Dropout(dropout_per)) # Dropout layer
model_reg.add(layers.BatchNormalization()) # Batch normalization layer
model_reg.add(layers.Dense(10, name='output',  activation='softmax'))
```

```python
model_reg.summary()
```

```python
learning_rate = 0.001
opt_reg = keras.optimizers.Adam(learning_rate = learning_rate)
loss_function = "categorical_crossentropy"
epochs = 25
batch_size = 512
model_reg.compile(optimizer = opt_reg, loss = loss_function, metrics = ["accuracy"])
history_reg = model_reg.fit(train_images,
                    train_labels,
                    epochs = epochs,
                    batch_size = batch_size,
                    validation_data = (val_images, val_labels),
                    callbacks=[early_stopping])
```

```python
history_dict_reg = history_reg.history
multiplot(history_reg)
```

```python
# Evaluate the model
predicted_values_reg = model_reg.predict(test_images)
predicted_classes_reg = np.argmax(predicted_values_reg, axis =1)
# Classification report for precision, recall and f1-score
report_reg = classification_report(test_labels, predicted_classes_reg, target_names = ['T-shirt/top', 
                                                                               'Trouser', 
                                                                               'Pullover', 
                                                                               'Dress', 
                                                                               'Coat', 
                                                                               'Sandal', 
                                                                               'Shirt', 
                                                                               'Sneaker', 
                                                                               'Bag', 
                                                                               'Ankle boot'])
# Print the evaluation metrics
print ("Classification Report:")
print (report_reg)
```
